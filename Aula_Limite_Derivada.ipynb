```python
import sympy as smp
import matplotlib.pyplot as plt
import numpy as np
x = smp.Symbol('x')
```

```python
f_a = (x**2 + x -2)/ (x-1)
print(smp.limit(f_a, x, 1))
smp.plot(f_a)
```

```python
f_b = (x+1)/(x-2)
print(smp.limit(f_b,x,2))
smp.plot(f_b, xlim=[-5,5],ylim=[-3,20])
```

```python
f_c = (2+3*x)
print(smp.limit(f_c,x,0))
smp.plot(f_c)
```

```python
f_d = (2+3*x)
print(smp.limit(f_d,x,smp.oo))
smp.plot(f_d)
```

```python
beta0 = smp.Symbol('beta0')
beta1 = smp.Symbol('beta1')
```

```python
f_e = (beta0+beta1*x)
print(smp.limit(f_e,x,0))
```

```python
f_f = (beta0+beta1*x)
print(smp.limit(f_f,x,smp.oo))
```

```python
f_g = (1/ 1+ smp.exp(-x))
print(smp.limit(f_g,x,smp.oo))
```

```python
f_h = (1/ 1+ smp.exp(-x))
print(smp.limit(f_h,x,-smp.oo))
```

# Constante


```python
funcao_original = x**2
funcao_derivada = smp.diff(funcao_original)
funcao_original = smp.lambdify(x,funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)

y = [funcao_original(i) for  i in range (-100,100)]
plt.plot(y)
plt.show()
y = [funcao_derivada(i) for i in range (-100,100)]
plt.plot(y)
plt.show()
```

# Potncia


```python
 funcao_b = 3*x**3 + 2*x**2 + x
funcao_derivada = smp.diff(funcao_b)
funcao_original_b = smp.lambdify(x,funcao_b)
funcao_derivada_b = smp.lambdify(x,funcao_derivada)

y = [funcao_original_b(i) for  i in range (-100,100)]
plt.plot(y)
plt.show()
y = [funcao_derivada_b(i) for i in range (-100,100)]
plt.plot(y)

plt.show()
```

# Multiplicao


```python
funcao_c = (x**2+3)*(2*x+1)
funcao_c = smp.diff(funcao_c)
funcao_c = smp.simplify(funcao_c)


funcao_original = funcao_c
funcao_derivada = smp.diff(funcao_c)
funcao_original = smp.lambdify(x, funcao_original)
funcao_derivada = smp.lambdify(x, funcao_derivada)
y = [funcao_original(i) for i in range (-100,100)]
plt.plot(y)
plt.show()
y = [funcao_derivada(i) for i in range (-100,100)]
plt.plot(y)
plt.show()
```

# Diviso

```python
funcao_d = (3*x**2+1)/(2*x-1)

funcao_d_derivada = smp.diff(funcao_d)
funcao_original = smp.lambdify(x,funcao_d)
funcao_d_derivada = smp.lambdify(x,funcao_d_derivada)
y = [funcao_original(i) for i in range (-10,10)]
plt.plot(y)
plt.show()
y = [funcao_d_derivada(i) for i in range (-100,100)]
plt.plot(y)
plt.show()
```

# Logaritmo

```python
 funcao_e = x * smp.log(x)
funcao_e_derivada = smp.diff(funcao_e)
funcao_original = smp.lambdify(x, funcao_e)
funcao_derivada = smp.lambdify(x, funcao_e_derivada)
y = [funcao_original(i) for i in range (1,10)]
plt.plot(y)
plt.show()
y = [funcao_derivada(i) for i in range (1,10)]
plt.plot(y)
plt.show()
```

# Exponencial

```python
funcao_f = (x*smp.exp(2*x))
funcao_f_derivada = smp.diff(funcao_f)
funcao_original = smp.lambdify(x, funcao_f)
funcao_derivada = smp.lambdify(x, funcao_f_derivada)
y = [funcao_original(i) for i in range (1,10)]
plt.plot(y)
plt.show()
y = [funcao_derivada(i) for i in range (1,10)]
plt.plot(y)
plt.show()
```

# Lucro


```python
p = smp.Symbol('p')
dev = (-2*p**2+50.4*p-10)
dev_d = smp.diff(dev)
funcao_original = smp.lambdify(p,dev)
funcao_derivada = smp.lambdify(p,dev_d)
y = [funcao_original(i) for i in range (20)]
plt.plot(y)
plt.show()
y = [funcao_derivada(i) for i in range (20)]
plt.plot(y)
plt.show()
lucro_maximo = np.max([funcao_original(i) for i in range(20)])
print('Lucro  R${:.2f}'.format(lucro_maximo))
```
